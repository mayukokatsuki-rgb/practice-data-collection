AMBIL DARI WEBSITE

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.yahoo.co.jp/"
html = requests.get(url)
soup = BeautifulSoup(html.content,"html.parser")

topics = soup.find_all(class_="_2j0udhv5jERZtYzddeDwcv")
for topic in topics:
    for element in topic.find_all("a"):
        print(element.text)
```

Summarize the Data

```python
import pandas as pd
df = pd.read_csv("test.csv")

print("数学の最高点=",df["数学"].max())
print("数学の最低点=",df["数学"].min())
print("数学の平均値=",df["数学"].mean())
print("数学の中央値=",df["数学"].median())
print("数学の合計=",df["数学"].sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
#名前の列をインデックスでCSVファイルを読み込む
df=pd.read_csv("test.csv",index_col=0)

#棒グラフ
df.plot.bar ()
plt.legend (loc="lower right")
plt.show ()

#棒グラフ (水平)
df.plot.barh ()
plt.legend (loc="lower left")
plt.show ()

#積み上げ棒グラフを使って表示
df.plot.bar (stacked=True)
plt.legend (loc="lower right")
plt.show ()

#箱ひげグラフを使って表示する
df.plot.box ()
plt.show ()

#面グラフを使って表示する
df.plot.area ()
plt.legend (loc="lower right")
plt.show ()
```

個別のデータをグラフで表示

```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
#名前の列をインデックスでCSVファイルを読み込む
df=pd.read_csv("test.csv",index_col=0)

#棒グラフ
df["国語"].plot.bar ()
plt.legend (loc="lower right")
plt.show ()

#棒グラフ (水平)
df[["国語","数学"]].plot.barh ()
plt.legend (loc="lower left")
plt.show ()

#棒グラフ (水平)
df.loc["C子"].plot.barh ()
plt.legend (loc="lower left")
plt.show ()

#円グラフ
df.loc["C子"].plot.pie (labeldistance=0.6)
plt.legend (loc="lower left")
plt.show ()
```
