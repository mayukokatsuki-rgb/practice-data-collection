```python
import yfinance as yf
import pandas as pd
import os

# List of stocks
stocks = ["BUMI.JK", "DEWA.JK", "BBRI.JK", "BBCA.JK", "ADRO.JK"]

all_data = []

for stock in stocks:

    ticker = yf.Ticker(stock)

    # Get today's data
    df = ticker.history(period="1d")

    df = df.reset_index()

    df["Ticker"] = stock

    # percentage change
    df["Pct_Change"] = ((df["Close"] - df["Open"]) / df["Open"]) * 100

    df = df[["Date","Ticker","Open","High","Low","Close","Pct_Change"]]

    all_data.append(df)

# combine all stocks
final_df = pd.concat(all_data)

# CSV filename
file_name = "stock_market_data.csv"

# If CSV exists → append
if os.path.exists(file_name):

    existing = pd.read_csv(file_name)

    final_df = pd.concat([existing, final_df])

    final_df.drop_duplicates(subset=["Date","Ticker"], inplace=True)

# save CSV
final_df.to_csv(file_name, index=False)

print("CSV database updated successfully")
```
